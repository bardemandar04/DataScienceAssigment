```python
#### Employee Management System

def Manu():
    emp_data={101:{"name":"Satya","age":27,"Department":"HR","Salary":50000}}
    temp = True
    while temp:
        print("Enter a number ")
        print("1 for Add new Employee details")
        print("2 for View all Employee")
        print("3 Search for Employee")
        num = int(input("4 for Exit"))
         
        if num==1:
            Add_Employee(emp_data)
            
        elif num == 2:
            View_Employees(emp_data)
        elif num == 3:
            Search_Employee(emp_data)
        elif num == 4:
            temp = False
        
            
    
def Add_Employee(Emp_Data):
    while True:
        ID = int(input("Enter ID"))
        if ID in Emp_Data:
            print("Inviled Id")
        else:
            name = input("Enter Name")
            age = int(input("Enter Age"))
            Department = input("Enter Department")
            salery = input("Enter Salery")
            Emp_Data[ID] = {
                "name":name,
                "age":age,
                "Department":Department,
                "Salary":salery
            }
            return Emp_Data
def View_Employees(Emp_Data):
    print("Emp_ID\tName\tAge\tDepartment\tSalary")
    for emp_id, details in Emp_Data.items():
        print(f"{emp_id}\t{details['name']}\t{details['age']}\t{details['Department']}\t\t{details['Salary']}")

def Search_Employee(Emp_Data):

    ID = int(input("Enter Employee ID"))
    print("Emp_ID\tName\tAge\tDepartment\tSalary")
    if(ID in Emp_Data):
        details = Emp_Data[ID]
        print(f"{ID}\t{details['name']}\t{details['age']}\t{details['Department']}\t\t{details['Salary']}")
          
    
    
Manu()
```
