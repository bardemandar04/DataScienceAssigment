### Basic Probality

#### a. Tossing a coin 10,000 times and calculating the experimental probability of heads and tails.

```python
import random as rd
Trail = 10000
head = 0
tail = 0
i = 0
for i in range(Trail):
    toss = rd.choice(['H','T'])
    if(toss == 'H'):
        head += 1
    else:
        tail += 1
print("Probality of heads is:",head/Trail)
print("Probality of tail is:",tail/Trail)
```

### Conditional Probability and Bayes' Theorem

#### 3. A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn randomly, its color noted, and it is put back into the bag.
#### If this process is repeated 1000 times, write a Python program to estimate:  
  a. The probability of drawing a red ball given that the previous ball was blue.  
  b. Verify Bayes' theorem with the simulation results.  


Steps  
    a. Use random sampling to simulate the process.  
    b. Compute conditional probabilities directly from the data.


```python
import random

# Define the bag
bag = ['R'] * 5 + ['G'] * 7 + ['B'] * 8

# Number of simulations
n_trials = 1000

# Store results
draws = []

# Simulate draws with replacement
for _ in range(n_trials):
    draws.append(random.choice(bag))

# Count occurrences
count_prev_blue = 0
count_red_given_prev_blue = 0

for i in range(1, n_trials):
    if draws[i - 1] == 'B':
        count_prev_blue += 1
        if draws[i] == 'R':
            count_red_given_prev_blue += 1
```

```python
# Estimated conditional probability
p_red_given_blue_sim = count_red_given_prev_blue / count_prev_blue

# Marginal probabilities from simulation
p_red_sim = draws.count('R') / n_trials
p_blue_sim = draws.count('B') / n_trials

# Bayes theorem estimate
p_blue_given_red_sim = p_blue_sim  # independence
p_red_given_blue_bayes = (p_blue_given_red_sim * p_red_sim) / p_blue_sim

print("Simulation Results:")
print(f"P(R | previous B) ≈ {p_red_given_blue_sim:.4f}")

print("\nBayes' Theorem Verification:")
print(f"P(R | B) using Bayes ≈ {p_red_given_blue_bayes:.4f}")

print("\nTheoretical Value:")
print("P(R | B) = 0.25")
```

### Random Variables and Discrete Probability
#### 

4. Generate a sample of size 1000 from a discrete random variable with the following distribution:  
  - P(X=1) = 0.25  
  - P(X=2) = 0.35  
  - P(X=3) = 0.4  
  Compute the empirical mean, variance, and standard deviation of the sample.  
  Steps  
      a. Use numpy.random.choice() to generate the sample.  
      b. Use numpy methods to calculate mean, variance, and standard deviation.


```python
import numpy as np
values = [1,2,3]
proba =[0.25,0.35,0.4]
sample = np.random.choice(values, size=1000, p=proba)
```

```python
mean = np.mean(sample)

# Empirical variance
variance = np.var(sample)       

# Empirical standard deviation
std_dev = np.std(sample)
print('mean of sample set is:',mean)
print('variance of samole set is', variance)
print('standead deviation of samole set is', std_dev)
```

### Continuous Random Variables
#### 

5. Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
  a. A histogram.  
  b. A probability density function (PDF) overlay.  
  Steps  
      a. Use numpy.random.exponential().  
      b. Use matplotlib to create visualizations. 


```python
import matplotlib.pyplot as plt
```

```python
mean = 5
Size = 2000
sample = np.random.exponential(scale=mean,size=Size)
```

```python
plt.figure(figsize=(8,5))
plt.hist(sample, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
```

```python
x = np.linspace(0, max(sample), 500)
pdf = (1 / mean) * np.exp(-x / mean)
plt.plot(x, pdf, 'r', linewidth=2, label='Exponential PDF (mean = 5)')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Exponential Distribution (Mean = 5)')
plt.legend()

# Show plot
plt.show()
```

### Central Limit Theorem#### 

6. Simulate the Central Limit Theorem by following these steps  
  a. Generate 10,000 random numbers from a uniform distribution.  
  b. Draw 1000 samples of size n = 30.  
  c. Calculate and visualize the distribution of sample means.  
  Steps  
      a. Use numpy.random.uniform().  
      b. Plot both the uniform distribution and the sample mean distribution for comparison.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility (optional)
np.random.seed(42)

# Step a: Generate 10,000 uniform random numbers
population = np.random.uniform(low=0, high=1, size=10000)

# Step b: Draw 1000 samples of size n = 30
n_samples = 1000
sample_size = 30

sample_means = []
for _ in range(n_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Step c: Visualization
plt.figure(figsize=(12, 5))

# Uniform distribution
plt.subplot(1, 2, 1)
plt.hist(population, bins=30, density=True, color='lightblue', edgecolor='black')
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

# Sample mean distribution
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, density=True, color='salmon', edgecolor='black')
plt.title('Distribution of Sample Means (n = 30)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

plt.tight_layout()
plt.show()

```
