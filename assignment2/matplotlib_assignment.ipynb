```python
import matplotlib.pyplot as plt
import random
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
```

```python
plt.figure()
plt.plot(x,y)
plt.title('line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
```

```python
plt.figure()
plt.bar(students,marks,color=['blue','red','orange','green'])
plt.title('students Marks graph')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
```

```python
explode =(0.1,0,0,0)
plt.figure()
plt.pie(revenue,labels=regions,explode=explode,autopct='%1.1f%%',startangle=90)
plt.title('Revenue distributation by region')
plt.show()
```

```python
sample_data = [random.randint(1, 100) for _ in range(1000)]
```

```python
plt.figure()
plt.hist(sample_data, bins=15, edgecolor='black') # You can pass the Pandas Series directly

# Add titles and labels
plt.title('Distributation of Random number')
plt.show()
```
