```python
import numpy as np
```

```python
arr1 = np.arange(1,21)
arr1
```

```python
sum= np.sum(arr1)
mean = np.mean(arr1)
mid = np.median(arr1)
std =np.std(arr1)
```

```python
print('sum of all number is:',sum)
print('mean is:',mean)
print('meadin of arr1 is:',mid)
print("Standard Deviation for arr1 is",std)
```

```python
indices_gt_10 = np.where(arr1 > 10)
print("Indices of elements > 10:", indices_gt_10)
```

```python
arr2 = np.arange(1,17).reshape(4,4)
arr2
```

```python
transpose_arr2 = arr2.T
transpose_arr2
```

```python
num_col_wise = np.sum(arr2,axis=0)
num_row_wise =np.sum(arr2,axis=1)
print("sum number in arr2 coloum wise",num_col_wise)
print("sum number in arr2 row wise",num_row_wise)
```

```python
arr3 = np.random.randint(1, 21, size=(3, 3))
arr4 = np.random.randint(1, 21, size=(3, 3))
print(arr3)
print(arr4)
```

```python
add = arr3+arr4
sub =arr3 - arr4
mul= arr3 * arr4
print("addition of two array is:")
print(add)
print("substraction of two array is")
print(sub)
print("multiplacition of two array is:")
print(mul)
```

```python
#dot product
dot_product = np.dot(arr3,arr4)
print("dot product of two array is")
print(dot_product)
```
