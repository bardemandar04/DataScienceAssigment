```python
import pandas as pd
```

```python
 data =pd.DataFrame({
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  })
```

```python
data.head()
```

```python
data.describe()
```

```python
avg_Salary = data[data['Department']=='HR']['Salary'].mean()
print("Avrage Salary of Employee from HR Department is:",avg_Salary)
```

```python
data['Bonus']=data['Salary']/10
data
```

```python
filtered_data = data[(data['Age']>25) & (data['Age']<30)]
filtered_data
```

```python
Avg_Salary_by_Dept = data.groupby('Department')['Salary'].mean()
Avg_Salary_by_Dept
```

```python
sorted_salary_data = data.sort_values(by='Salary', ascending=False)
sorted_salary_data.to_csv("salary sorted data")
```
