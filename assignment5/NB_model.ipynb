```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.impute import SimpleImputer # To handle missing values
```

```python
data = pd.read_csv('Titanic-Dataset.csv')
data.head()
```

```python
# Select and Prepare Features
data_model = data[['Survived','Pclass','Sex','Age','Fare']]


#converting sex 0 as male and 1 as female

data_model['Sex'] = data_model['Sex'].map({'male':0, 'female':1})
```

```python
data_model.head()
```

```python
data_model.isnull().sum()
```

```python
# handling missing values 

# separate feature (x) and target (y)

X = data_model.drop('Survived',axis =1)
Y = data_model['Survived']

```

```python
imputer = SimpleImputer(strategy = 'mean')
X_imputed = imputer.fit_transform(X)
```

```python
X_imputed[:10]
```

```python
np.isnan(X_imputed).any()
```

```python
# spliting data into traning and testing

X_train,X_test,Y_train,y_test=train_test_split(X_imputed,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape,Y_train.shape
```

```python
# training model

nb_model = GaussianNB()

# training Model
nb_model.fit(X_train,Y_train)
```

```python
y_prad = nb_model.predict(X_test)
```

```python
# Evaluating model
print("accuracy:", accuracy_score(y_test,y_prad))
print("\n Confusion Matric\n", confusion_matrix(y_test,y_prad))
print("Classification Report:\n", classification_report(y_test,y_prad))
```
